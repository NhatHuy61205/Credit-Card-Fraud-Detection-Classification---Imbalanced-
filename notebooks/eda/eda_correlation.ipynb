### EDA_CORRELATION

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../../data/creditcard.csv")
df.head()
```

### Ma trận tương quan

```python
corr = df.corr()
corr.head()
```

### Biểu đồ Heatmap

```python
plt.figure(figsize=(12,10))
sns.heatmap(df.corr(), cmap='coolwarm')
plt.show()
```

Biểu đồ heatmap thể hiện ma trận tương quan giữa các biến trong bộ dữ liệu. Có thể quan sát thấy các biến V1–V28 gần như không tương quan với nhau, đảm bảo tính độc lập tuyến tính giữa các đặc trưng.

Không có đa cộng tuyến nghiêm trọng.

Nguyên nhân là do các biến V1–V28 đã được biến đổi bằng PCA, do đó chúng gần như trực giao với nhau.

### Tương quan với biến mục tiêu (Class)

```python
corr_class = corr['Class'].sort_values(ascending=False)
print(corr_class)
```

Các biến quan trọng

Các biến V17, V14 và V12 có hệ số tương quan âm cao nhất (|corr| > 0.25), cho thấy chúng có mối liên hệ mạnh với giao dịch gian lận.

Điều này cho thấy: Khi giá trị các biến này giảm, khả năng xảy ra gian lận có xu hướng tăng

### Biểu đồ phân tán của 2 class quan trọng

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df,
                x='V17',
                y='V14',
                hue='Class',
                alpha=0.6)

plt.title("Phân tán giữa V17 và V14 theo Class")
plt.show()
```

Biểu đồ phân tán giữa V17 và V14 cho thấy sự tách biệt tương đối rõ ràng giữa giao dịch hợp lệ và giao dịch gian lận.

Các giao dịch gian lận chủ yếu tập trung ở vùng giá trị âm lớn của cả hai biến.
Điều này cho thấy V14 và V17 là những đặc trưng quan trọng trong việc phát hiện gian lận.
