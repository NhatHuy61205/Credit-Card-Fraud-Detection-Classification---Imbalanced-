### EDA_CORRELATION

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../../data/creditcard.csv")
df.head()
```

### Ma trận tương quan

```python
corr = df.corr()
corr.head()
```

### Tương quan với biến mục tiêu

```python
corr_class = corr['Class'].sort_values(ascending=False)
print(corr_class)
```

Ta có thể thấy được 6 biến có mức độ tương quan cao nhất với biến mục tiêu Class là:
V17, V14, V12, V7, V11, V4

V17, V14, V12, V7 tương quan âm với biến mục tiêu

V11, V4 tương quan dương

Cho thấy chúng có mối liên hệ mạnh với giao dịch gian lận

### Biểu đồ Heatmap

```python
cols = ['Class','V17','V14','V12','V7','V11','V4']
corr = df[cols].corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, 
            annot=True,      
            cmap='coolwarm',
            fmt=".2f")

plt.title("Heatmap của 6 biến quan trọng")
plt.show()
```

Các biến này gần như không tương quan với nhau(các giá trị gần 0.00 giữa các cặp biến), cho thấy không có hiện tượng đa cộng tuyến đáng kể.

Nguyên nhân là do các biến đã được biến đổi bằng PCA.

### Biểu đồ phân tán

```python
sns.lmplot(
    data=df,
    x='V17',
    y='V14',
    hue='Class',
    height=6,
    aspect=1.3,
    scatter_kws={'alpha':0.6}
)

plt.title("Phân tán giữa V17 và V14")
plt.show()
```

```python
sns.lmplot(
    data=df,
    x='V17',
    y='V12',
    hue='Class',
    height=6,
    aspect=1.3,
    scatter_kws={'alpha':0.6}
)

plt.title("Phân tán giữa V17 và V12")
plt.show()
```

Các biểu đồ phân tán trên cho thấy sự tách biệt tương đối rõ ràng giữa giao dịch hợp lệ và giao dịch gian lận.

Các giao dịch gian lận chủ yếu tập trung ở vùng giá trị âm lớn của cả hai biến.

```python
sns.lmplot(
    data=df,
    x='Amount',
    y='Time',
    hue='Class',
    height=6,
    aspect=1.3,
    scatter_kws={'alpha':0.6}
)

plt.show()
```

Phần lớn các điểm tập trung ở Amount nhỏ.

Điều này cho thấy đa số giao dịch có giá trị thấp.

Các giao dịch gian lận xuất hiện rất ít so với giao dịch bình thường, cho thấy dữ liệu bị mất cân bằng. 

Đường hồi quy có xu hướng dốc xuống nhưng các điểm dữ liệu phân tán rộng, cho thấy mối tương quan giữa Amount và Time là khá yếu.
